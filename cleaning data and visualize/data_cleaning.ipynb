```python
def clean_price_column(col):
    # delete "تومان"
    col = col.str.replace("تومان", "")
    # make , to ""
    col = col.str.replace("[٬, ]", "", regex=True)
    return col

df["price"] = clean_price_column(df["price"])
df["price_per_meter"] = clean_price_column(df["price_per_meter"])
df.head()
```

```python
df["price"] = df["price"].astype("Int64")
df.head()
```

```python
mask = ~df["price_per_meter"].str.match(r'^\d+$')

df[mask]
```

```python
df = df[~mask]
```

```python
df["price_per_meter"] = df["price_per_meter"].astype("Int64")
df.head()
```

```python
import re
import numpy as np
```

```python
def convert_floor(value):
    value = str(value).strip()

    # persian to english
    persian_digits = '۰۱۲۳۴۵۶۷۸۹'
    english_digits = '0123456789'
    trans_table = str.maketrans(persian_digits, english_digits)
    value = value.translate(trans_table)

    if 'همکف' in value and 'زیر' not in value:
        return 0
    elif 'زیرهمکف' in value or 'زیر' in value:
        return -1
    elif re.match(r'^\d+ از \d+$', value): #number from total floors
        return int(value.split(' از ')[0])
    elif value.isdigit():  #just one number
        return int(value)
    else:
        return np.nan  # if any other value

df["floor_number"] = df["floor"].apply(convert_floor)
df.head()
```

```python
df = df.drop(columns=["floor"])
df.head()
```

```python
def extract_area(title):
    match = re.search(r'(\d+)\s*متر', title)  # مثل "165 متر" یا "۸۷ متر"
    if match:
        return int(match.group(1))
    
    match_alt = re.search(r'(\d+)\s*متری', title)  # مثل "۵۳ متری"
    if match_alt:
        return int(match_alt.group(1))
    
    return None

df["area"] = df["title"].apply(extract_area)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

top_locations = df["location_encoded"].value_counts().head(20).index

df["location_top"] = df["location_encoded"].apply(lambda x: x if x in top_locations else -1)

plt.figure(figsize=(14, 8))
scatter = sns.scatterplot(
    data=df,
    x="area",
    y="price",
    hue="location_top",
    palette="tab20",
    alpha=0.6,
    edgecolor="w",
    s=60
)

plt.title("َArea & Location & Price", fontsize=16)
plt.xlabel("Arrea", fontsize=13)
plt.ylabel("Price", fontsize=13)
plt.legend(title="encoded location", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
#delete outliner datas for better plot
Q1 = df["price"].quantile(0.25)
Q3 = df["price"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_filtered = df[(df["price"] >= lower_bound) & (df["price"] <= upper_bound)]
```

```python
import seaborn as sns

sample_df = df_filtered.sample(n=2000, random_state=42)

plt.figure(figsize=(14,7))
sns.scatterplot(
    data=sample_df,
    x="location_encoded",
    y="price",
    hue="location_encoded",
    palette="tab20",
    alpha=0.6,
    s=40,
    legend=None
)
plt.title("Scatter plot (random sample of 2000 listings)")
plt.xlabel("Location (encoded)")
plt.ylabel("Price (Toman)")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
